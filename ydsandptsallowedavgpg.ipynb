```python
import nfl_data_py as nfl
```

```python
import pandas as pd
```

```python
weekly_nfl_data = nfl.import_weekly_data([2024])

weekly_nfl_data = weekly_nfl_data[weekly_nfl_data['week'] != 5]

#weekly_nfl_data

#print(max(weekly_nfl_data['week']))
```

```python
rb_only = weekly_nfl_data[weekly_nfl_data['position'] == 'RB']


rushing_yards_pg_only_rb = rb_only.groupby(['opponent_team', 'week'], as_index=False)['rushing_yards'].sum()


AVG_rushing_yards_pg_only_rb = rushing_yards_pg_only_rb.groupby(['opponent_team'], as_index=False)['rushing_yards'].mean()

AVG_rushing_yards_pg_only_rb.rename(columns={'rushing_yards': 'AVG_rush_yds_allowed_pg_RB', 'opponent_team': 'team'}, inplace=True)

AVG_rushing_yards_pg_only_rb = AVG_rushing_yards_pg_only_rb.sort_values(by='AVG_rush_yds_allowed_pg_RB', ascending=False)

#AVG_rushing_yards_pg_only_rb
```

```python
receiving_yards_pg_only_rb = rb_only.groupby(['opponent_team', 'week'], as_index=False)['receiving_yards'].sum()


AVG_receiving_yards_pg_only_rb = receiving_yards_pg_only_rb.groupby(['opponent_team'], as_index=False)['receiving_yards'].mean()

AVG_receiving_yards_pg_only_rb.rename(columns={'receiving_yards': 'AVG_receiving_yds_allowed_pg_RB', 'opponent_team': 'team'}, inplace=True)

AVG_receiving_yards_pg_only_rb = AVG_receiving_yards_pg_only_rb.sort_values(by='AVG_receiving_yds_allowed_pg_RB', ascending=False)

#AVG_receiving_yards_pg_only_rb
```

```python
FINAL_DF = pd.merge(AVG_rushing_yards_pg_only_rb, AVG_receiving_yards_pg_only_rb, how='left', on='team')

FINAL_DF
```

```python
#Avg fantasy points allowed for each position per game

fantasy_points_each_game = weekly_nfl_data.groupby(['opponent_team', 'week', 'position'], as_index=False)['fantasy_points_ppr'].sum()

positions_to_filter = ['RB', 'WR', 'TE', 'QB']

fantasy_points_each_game = fantasy_points_each_game[fantasy_points_each_game['position'].isin(positions_to_filter)]

AVG_fantasy_points_each_game_ALL_POS = fantasy_points_each_game.groupby(['opponent_team', 'position'], as_index=False)['fantasy_points_ppr'].mean()

AVG_fantasy_points_each_game_ALL_POS
```

```python
AVG_fantasy_points_each_game_ALL_POS_pivot = AVG_fantasy_points_each_game_ALL_POS.pivot(index='opponent_team', columns='position', values='fantasy_points_ppr').reset_index()

AVG_fantasy_points_each_game_ALL_POS_pivot.rename(columns={'opponent_team': 'team', 'QB': 'AVG_fantasy_points_allowed_pg_QB', 'RB': 'AVG_fantasy_points_allowed_pg_RB', 'WR': 'AVG_fantasy_points_allowed_pg_WR', 'TE': 'AVG_fantasy_points_allowed_pg_TE'}, inplace=True)

#AVG_fantasy_points_each_game_ALL_POS_pivot = AVG_fantasy_points_each_game_ALL_POS_pivot.drop('position', axis=1)

#AVG_fantasy_points_each_game_ALL_POS_pivot
```

```python
FINAL_DF = pd.merge(FINAL_DF, AVG_fantasy_points_each_game_ALL_POS_pivot, how='left', on='team')

FINAL_DF
```
